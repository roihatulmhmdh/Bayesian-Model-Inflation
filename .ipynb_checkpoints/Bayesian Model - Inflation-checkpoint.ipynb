# Background 
---

Inflasi adalah kenaikan harga barang dan jasa secara umum dan terus menerus dalam jangka waktu tertentu. Perhitungan inflasi dilakukan oleh Badan Pusat Statistik (BPS) di Indonesia. BPS melakukan survei untuk mengumpulkan data harga dari berbagai macam barang dan jasa yang dianggap mewakili belanja konsumsi masyarakat. Data tersebut kemudian digunakan untuk menghitung tingkat inflasi dengan membandingkan harga-harga saat ini dengan periode sebelumnya.

Akan dilakukan prediksi inflasi dengan memperhatikan faktor-faktor lain.

# Dataset Description : 
---

- `date` : tanggal (frekuensi bulanan)
- `inflation` : Tingkat inflasi di Indonesia
- `birate` : Tingkat suku bunga Bank Indonesia
- `ihs` : Indeks Harga Saham Komposit di Bursa Efek

# Initial Model 
---

Diawali dengan model yang sederhana, dengan hanya menggunakan variabel BI Rate

**Simple Model**

$\text{inflation} =  \alpha + \beta * birate$

## Prior Selection  Predictive Checking
---

```python
# Visualize the distribution of the target variable
plt.figure(figsize=(6, 4))
sns.regplot(x=df['birate'], y=df['inflation'])
plt.title('Distribution of Inflation')
plt.show()
```

Berdasarkan gambar diatas, dapat kita lihat bahwa slope berada dalam rentang 1 hingga 2, maka dapat ditentukan prior sebagai berikut:




**Model**:


$$
\begin{align}
\text{inflation} =  \alpha + \beta * birate
\end{align}
$$


**Distribution**:


$$
\begin{align}
\text{Inflation}_i & \sim Normal(\mu_i,\sigma) \text{ [likelihood]} \\
\mu_i &= \alpha + \beta \cdot \text{BI Rate}_i \\ 
\alpha & \sim Normal(\mu = 2,\sigma =1) \text{ [prior]} \\
\beta & \sim Normal(\mu = 2,\sigma =1) \text{ [prior]} \\
\sigma & \sim Expon(\lambda = 2)\text{ [prior]} \\
\end{align}
$$


```python
with pm.Model() as inf_1 : 
    # prior 
    a = pm.Normal('a',2,1)
    b = pm.Normal('b',2,1)
    sigma = pm.Exponential('sigma',2)
    
    # likelihood
    linear_model = pm.Deterministic('mu',a + b * df['birate'].values)
    inf = pm.Normal('inflation',mu=linear_model,sigma=sigma,observed=df['inflation'].values)
```

```python
# sample from prior distribution 
with inf_1 :
      inf_1_prior = pm.sample_prior_predictive()
```

```python
shape_prior_samples = inf_1_prior.prior['a'].values.flatten().shape[0]
n_data = df.shape[0] 
pred_inf = np.zeros(( n_data , shape_prior_samples ))


# predicting priors using intercept + coefficient obtained from prior sampling
for i, val in enumerate(df['birate']):
    pred_inf[i] = inf_1_prior.prior["a"] + inf_1_prior.prior["b"]  * val


fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(10,5))
for i in range(shape_prior_samples) : 
    ax.plot(df['birate'],pred_inf[:,i], c="k", alpha=0.4)

ax.scatter(df['birate'],df['inflation'], c="blue", alpha=0.6)
ax.set_title('Prior Prediction Plot')
```

Karena model belum melihat data, kisaran koefisiennya agak melenceng

## Fitting Model 

---

```python
with inf_1 :
      inf_1_trace = pm.sample(2000,random_seed=72,target_accept=0.9)
    
az.summary(inf_1_trace,var_names=["~mu"])
```

## Checking Sampling Process 

---

```python
az.plot_trace(inf_1_trace,var_names=["~mu"])
```

**Gambar Kiri**: Density posterior. Terdapat beberapa garis yang berasal dari chain  di pymc. Istilah chain  mengacu pada proses pengambilan sampel independen 

**Gambar Kanan**: Perubahan nilai parameter selama proses pengambilan sampel posterior 

Kondisi baik selama pengambilan sampel posterior, dilihat dari nilai parameter $\alpha$, $\beta$, dan $\sigma$ menuju nol, atau tidak bertambah/berkurang terlalu banyak.

```python
az.summary(inf_1_trace,var_names=["~mu"])
```

Model bekerja dengan baik dalam proses pengambilan sampel, dilihat dari Rhat dan Effective Sample Size.

```python
az.plot_autocorr(inf_1_trace,var_names=['a','b','sigma'])
```

Autocorrelation berhubungan dengan Effective Sample Size, memiliki low autocorrelation. Namun terlihat $\alpha$ and $\beta$ memiliki high correlation pada awal iterasi.

# Updating Model : Adding More Features
--- 

## Prior Selection  Predictive Checking
---

**Model**:


$$
\begin{align}
\text{inflation} =  \alpha + \beta_{1} * birate + \beta_{2} * ihs
\end{align}
$$


**Distribution**:


$$
\begin{align}
\text{Inflation}_i & \sim Normal(\mu_i,\sigma) \text{ [likelihood]} \\
\mu_i &= \alpha + \beta_{1} \cdot \text{BI Rate}_i + \beta_{2} \cdot \text{IHS}_i \\ 
\alpha & \sim Normal(\mu = 0,\sigma =1) \text{ [prior]} \\
\beta_{1} & \sim Normal(\mu = 2,\sigma =1) \text{ [prior]} \\
\beta_{2} & \sim Normal(\mu = 2,\sigma =1) \text{ [prior]} \\
\sigma & \sim Expon(\lambda = 2)\text{ [prior]} \\
\end{align}
$$



```python
with pm.Model() as inf_2 : 
    # prior 
    a = pm.Normal('a',0,1)
    b1 = pm.Normal('b_birate',2,1)
    b2 = pm.Normal('b_ihs',2,1)
    sigma = pm.Exponential('sigma',2)
    
    # likelihood
    linear_model = pm.Deterministic('mu', a + b1 * df['birate'].values \
                                   + b2 * df['ihs'].values  )
    inf = pm.Normal('inflation',mu=linear_model,sigma=sigma,observed=df['inflation'].values)


# sample from prior distribution 
with inf_2 :
      inf_2_prior = pm.sample_prior_predictive(random_seed=72)
```

```python
shape_prior_samples = inf_2_prior.prior['a'].values.flatten().shape[0]
n_data = df.shape[0] 
pred_inf = np.zeros(( n_data , shape_prior_samples ))


# predicting priors using intercept + coefficient obtained from prior sampling
for idx in range(df.shape[0]) :
    pred_inf[idx] = inf_2_prior.prior["a"] + inf_2_prior.prior["b_birate"]  * df['birate'].values[idx] \
                    +  inf_2_prior.prior["b_ihs"]  * df['ihs'].values[idx]
```

```python
index_ = df['birate'].values.argsort()
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(12,4))


for idx in range(2) : 
    ax[0].scatter(df['birate'].values[index_],pred_inf[index_,idx], c="k", alpha=0.4)
    
ax[0].scatter(df['birate'].values[index_],df['inflation'].values[index_], c="blue", alpha=0.6)

index_ihs = df['ihs'].values.argsort()

for idx in range(2) : 
    ax[1].scatter(df['ihs'].values[index_ihs],pred_inf[index_ihs,idx], c="k", alpha=0.4)
    
ax[1].scatter(df['ihs'].values[index_ihs],df['inflation'].values[index_ihs], c="blue", alpha=0.6)
```

# Updating Model :  Changing the Likelihood 
---

## Prior Selection  Predictive Checking
---

**Model**:


$$
\begin{align}
\text{inflation} =  \alpha + \beta_{1} * birate + \beta_{2} * ihs
\end{align}
$$


**Distribution**:


$$
\begin{align}
\text{Inflation}_i & \sim Student-t(\mu_i,\sigma, n) \text{ [likelihood]} \\
\mu_i &= \alpha + \beta_{1} \cdot \text{BI Rate}_i + \beta_{2} \cdot \text{IHS}_i \\ 
\alpha & \sim Normal(\mu = 0,\sigma =1) \text{ [prior]} \\
\beta_{1} & \sim Normal(\mu = 2,\sigma =1) \text{ [prior]} \\
\beta_{2} & \sim Normal(\mu = 2,\sigma =1) \text{ [prior]} \\
\sigma & \sim Expon(\lambda = 1)\text{ [prior]} \\
n & \sim Uniform(\mu = 1,\sigma =3)\text{ [prior]} \\
\end{align}
$$



```python
with pm.Model() as inf_3 : 
    # prior 
    a = pm.Normal('a',0,1)
    b1 = pm.Normal('b_birate',2,1)
    b2 = pm.Normal('b_ihs',2,1)
    sigma = pm.Exponential('sigma',2)
    nu = pm.Uniform('nu',1,3)
    
    # likelihood
    linear_model = pm.Deterministic('mu', a + b1 * df['birate'].values \
                                   + b2 * df['ihs'].values )
    inf = pm.StudentT('inflation',nu=nu,mu=linear_model,sigma=sigma,observed=df['inflation'].values)


# sample from prior distribution 
with inf_3 :
      inf_3_prior = pm.sample_prior_predictive(random_seed=72)
```

```python
shape_prior_samples = inf_3_prior.prior['a'].values.flatten().shape[0]
n_data = df.shape[0] 
pred_inf = np.zeros(( n_data , shape_prior_samples ))


# predicting priors using intercept + coefficient obtained from prior sampling
for idx in range(df.shape[0]) :
    pred_inf[idx] = inf_3_prior.prior["a"] + inf_3_prior.prior["b_birate"]  * df['birate'].values[idx] \
                    +  inf_3_prior.prior["b_ihs"]  * df['ihs'].values[idx]
```

```python
index_ = df['birate'].values.argsort()
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(12,4))


for idx in range(2) : 
    ax[0].scatter(df['birate'].values[index_],pred_inf[index_,idx], c="k", alpha=0.4)
    
ax[0].scatter(df['birate'].values[index_],df['inflation'].values[index_], c="blue", alpha=0.6)

index_ihs = df['ihs'].values.argsort()

for idx in range(2) : 
    ax[1].scatter(df['ihs'].values[index_ihs],pred_inf[index_ihs,idx], c="k", alpha=0.4)
    
ax[1].scatter(df['ihs'].values[index_ihs],df['inflation'].values[index_ihs], c="blue", alpha=0.6)
```

# Changing into Quadratic Form
---

## Prior Selection  Predictive Checking
---

**Model**:


$$
\begin{align}
\text{inflation} =  \alpha + \beta_{1} * birate + \beta_{quad} * birate ^2
\end{align}
$$


**Distribution**:


$$
\begin{align}
\text{Inflation}_i & \sim Normal(\mu_i,\sigma) \text{ [likelihood]} \\
\mu_i &= \alpha + \beta_{1} \cdot \text{BI Rate}_i + \beta_{2} \cdot \text{IHS}_i \\ 
\alpha & \sim Normal(\mu = 0,\sigma =1) \text{ [prior]} \\
\beta_{1} & \sim Normal(\mu = 2,\sigma =1) \text{ [prior]} \\
\beta_{2} & \sim Normal(\mu = 2,\sigma =1) \text{ [prior]} \\
\sigma & \sim Expon(\lambda = 1)\text{ [prior]} \\
\end{align}
$$



```python
with pm.Model() as inf_4 : 
    # prior 
    a = pm.Normal('a',0,1)
    b1 = pm.Normal('b_birate',2,1)
    b_quad = pm.Normal('b_birate_sq',2,1)
    sigma = pm.Exponential('sigma',1)
    
    # likelihood
    
    linear_model = pm.Deterministic('mu', a + b1 * df['birate'].values \
                                   + b_quad * (df['birate'].values**2)  )

    power = pm.Normal('inflation',mu=linear_model,sigma=sigma,observed=df['inflation'].values)

with inf_4 :
      inf_4_prior = pm.sample_prior_predictive(random_seed=72)
```

```python
shape_prior_samples = inf_4_prior.prior['a'].values.flatten().shape[0]
n_data = df.shape[0] 
pred_power = np.zeros(( n_data , shape_prior_samples ))


# predicting priors using intercept + coefficient obtained from prior sampling
# sort values based on 
for idx in range(df.shape[0]) :
    pred_power[idx] = inf_4_prior.prior["a"] + inf_4_prior.prior["b_birate"]  * df['birate'].values[idx] \
                    +  inf_4_prior.prior["b_birate_sq"]  * df['birate'].values[idx]**2
```

```python
index_ = df['birate'].values.argsort()
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(12,4))

for idx in range(100) : 
    ax[0].scatter(df['birate'].values[index_],pred_power[index_,idx], c="k", alpha=0.4)
    
ax[0].scatter(df['birate'].values[index_],df['inflation'].values[index_], c="blue", alpha=0.6)
ax[0].set_title('Prior Prediction')



index_br = (df['birate'].values**2).argsort()

for idx in range(100) : 
    ax[1].scatter(df['birate'].values[index_br]**2,pred_power[index_br,idx], c="k", alpha=0.4)
    
ax[1].scatter(df['birate'].values[index_br],df['inflation'].values[index_br], c="blue", alpha=0.6)
ax[1].set_title('Prior Prediction Plot')
```
